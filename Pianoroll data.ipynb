```python
import pypianoroll
import matplotlib.pyplot as plt
```

```python
multitrack = pypianoroll.load("20d47f4c18db82fd2a3f462658019663.npz")
print(multitrack)
print("=============================================================================================")
# general file data
print("Multitrack General Data:")
print(f"Name: {multitrack.name}")
print(f"Resolution: {multitrack.resolution}")
print(f"Tempo array size: {multitrack.tempo.shape[0]}")
print(f"Downbeat array size: {multitrack.downbeat.shape[0]}")
print()

# each track information
print("Tracks Information:")
for track in multitrack.tracks:
    print(f"Track Name: {track.name}")
    print(f"Program: {track.program}")
    print(f"Is Drum: {track.is_drum}")
    print(f"Pianoroll shape: {track.pianoroll.shape}")
    print()
```

```python
plt.figure(figsize=(18, 20))  
multitrack.plot()
plt.show()
```
